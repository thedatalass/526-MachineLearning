```python
#From Jake VanderPlas
#https://jakevdp.github.io/blog/2017/03/03/reproducible-data-analysis-in-jupyter/
URL = 'https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD'
```

```python
from urllib.request import urlretrieve
urlretrieve(URL, 'Fremont.csv')
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
data = pd.read_csv('Fremont.csv', index_col='Date', parse_dates=True)
data.head()
```

```python
#Put plot inside Notebook
%matplotlib inline
data.plot()
```

```python
#This data's a bit messy so let's just look at it by week
%matplotlib inline
data.resample('W').sum().plot();
```

```python
#Change plot style
import matplotlib.pyplot as plt
%matplotlib inline
data.columns = ['West', 'East']
data.resample('W').sum().plot();
```

```python
#Let's look at the annual trends
data.resample('D').sum().rolling(365).sum().plot();
```
